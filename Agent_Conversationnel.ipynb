**2. Utilisation de LightRAG afin d'interroger le système**

```python
import nest_asyncio
nest_asyncio.apply()

!curl -sSL https://ollama.com/install.sh | bash

!ollama start > /dev/null 2>&1 &

!ollama pull gemma2:2b

!ollama pull "nomic-embed-text"

from lightrag import LightRAG,QueryParam
from lightrag.llm.ollama import ollama_model_complete, ollama_embed
from lightrag.utils import EmbeddingFunc

WORKING_DIR = "/content/test_rag"

class CustomRAG(LightRAG):
    def _generate_prompt(self, query: str, context: str) -> str:
        return f"Utilisez UNIQUEMENT les informations suivantes pour répondre : {context}\nQuestion : {query}\nRéponse basée uniquement sur les informations ci-dessus :"

# Utilisation de la classe personnalisée
rag = CustomRAG(
    working_dir=WORKING_DIR,
    llm_model_func=ollama_model_complete,
    llm_model_name='gemma2:2b',
    llm_model_max_async=4,
    llm_model_max_token_size=32768,
    llm_model_kwargs={"host": "http://localhost:11434/", "options": {"num_ctx": 32768}},
    embedding_func=EmbeddingFunc(
        embedding_dim=768,
        max_token_size=8192,
        func=lambda texts: ollama_embed(
            texts,
            embed_model="nomic-embed-text",host="http://localhost:11434/"
        )
    ),
)


rag_passable = CustomRAG(
    working_dir=WORKING_DIR,
    llm_model_func=ollama_model_complete,
    llm_model_name='gemma2:2b',
    llm_model_max_async=4,
    llm_model_max_token_size=32768,
    llm_model_kwargs={"host": "http://localhost:11434/", "options": {"num_ctx": 32768}},
    embedding_func=EmbeddingFunc(
        embedding_dim=768,
        max_token_size=8192,
        func=lambda texts: ollama_embed(
            texts,
            embed_model="nomic-embed-text",host="http://localhost:11434/"
        )
    ),
)

rag_mediocre = CustomRAG(
    working_dir=WORKING_DIR,
    llm_model_func=ollama_model_complete,
    llm_model_name='gemma2:2b',
    llm_model_max_async=4,
    llm_model_max_token_size=32768,
    llm_model_kwargs={"host": "http://localhost:11434/", "options": {"num_ctx": 32768}},
    embedding_func=EmbeddingFunc(
        embedding_dim=768,
        max_token_size=8192,
        func=lambda texts: ollama_embed(
            texts,
            embed_model="nomic-embed-text",host="http://localhost:11434/"
        )
    ),
)

#Insertion des data pour répondre à la question
#Documents à insérer
documents = [
    "L'IUT de Vannes est située dans la ville de Vannes, en Bretagne.",
    "L'IUT de Vannes fait partie de l'université bretagne sud (UBS).",
    "L'IUT de Vannes comporte quatre département : Informatique (INFO), Gestion et Administration des entreprises (GEA), Science des données (SD) et Techniques de communication (TC)",
    "La formation de l'IUT, le BUT dure trois ans",
    "Durant ces trois années, il est possible de réaliser un semestre à l'étranger ou une alternance",
    "L'IUT de Vannes a été crée en 1970",
    "Son directeur actuel est Xavier Tripoteau"
]

documents_mediocres = [
    "L'IUT de Vannes est située dans la ville de Vannes.",
    "L'IUT de Vannes fait partie de l'université bretagne sud (UBS).",
    "L'IUT de Vannes comporte quatre département.",
    "La formation de l'IUT est le BUT",
]

documents_passables = [
    "L'IUT de Vannes est située dans la ville de Vannes, en Bretagne.",
    "L'IUT de Vannes fait partie de l'université bretagne sud (UBS).",
    "L'IUT de Vannes comporte quatre département : INFO, GEA, SD et TC",
    "La formation de l'IUT, le BUT dure trois ans",
    "Durant ces trois années, il est possible de réaliser un semestre à l'étranger",
    "Son directeur actuel est Xavier Tripoteau"
]

#Insertion des documents
rag.insert(documents)
rag_mediocre.insert(documents_mediocres)
rag_passable.insert(documents_passables)

#Affichage du résultat
result_bad = rag_mediocre.query(query_bad, param=QueryParam(mode="local"))
result_mid = rag_passable.query(query_mid, param=QueryParam(mode="local"))
result_good = rag.query(query_good, param=QueryParam(mode="local"))

print("result_bad : "+result_bad)
print("result_mid : "+result_mid)
print("result_good :"+result_good)
```

**3.Génération de l'audio de réponse avec une voix personnalisée**

Configuration 1 - Mauvaise qualité

```python
from TTS.api import TTS

# Initialisation du modèle XTTS v2
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2", gpu=True)

# Génération de la synthèse vocale
tts.tts_to_file(
    text=result_bad,
    file_path="output_bad.wav",
    speaker_wav="speaker_bad.wav",
    language="fr"
)
```
